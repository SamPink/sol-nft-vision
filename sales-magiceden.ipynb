```python
import http.client
import pandas as pd
import time
```

```python
def get_request(slug, page):
    conn = http.client.HTTPSConnection("api-mainnet.magiceden.dev")
    payload = ''
    headers = {}
    offset = page * 100
    conn.request("GET", f"/v2/collections/{slug}/activities?offset={offset}&limit=100&type=buyNow", payload, headers)
    res = conn.getresponse()
    data = res.read()
    data = data.decode("utf-8")
    return data
```

```python

#okay_bears = "12f5732ce6ca33cbe4091cefbebf93543245e0278b4a557e5204e3d76287b289"
page_num = 0
search = True
all_trades= pd.DataFrame()
while search:
    try:
        sales_page = get_request(slug="okay_bears", page=page_num)

        df = pd.read_json(sales_page)
        
        all_trades = all_trades.append(df)

        page_num += 1
        print(f"Page {page_num}")
        time.sleep(1)
    except Exception as e:
        print(e)
        search = False

    # limit to 100 calls
    if page_num == 300:
        search = False
```

```python
all_trades.drop_duplicates(['blockTime', 'price']).to_csv("data/okay_bear/transactions.csv")
```
